### 問1

```python
import warnings
warnings.filterwarnings("ignore", message="numpy.dtype size changed")
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")

import numpy as np
import matplotlib.pyplot as plt
import cis
```

```python
def curve(a, f, fs=8000, sec=1):
    """
    sec: サンプリング秒数
    """
    r = np.arange(0,sec,1/fs)
    return a*np.sin(2*np.pi*f*r),r

# (1)
fs = 8000
a = 1
f1 = 262
y1,range1 = curve(a,f1,sec=0.5)

plt.plot(range1[:100], y1[:100])
plt.show()
cis.audioplay(y1, fs)
```

```python
# (2)
a = 1
f2=440
y2,range2 = curve(a,f2,sec=2)
plt.plot(range2[:100], y2[:100])
plt.show()
cis.audioplay(y2, fs)
```

### 問3
問題の都合上, 先に問3を解く.

#### (1)
出力するサンプリング点数は, サンプリング周波数 $fs$ なので $0.05*fs$ サンプリング回数 となる.

```python
sec = 0.05
l2 = int(sec*fs)
cis.audioplay(y1[:l2], fs)
```

#### (2)
1周期は $1/f$ 秒 なので, これのサンプリング点数は $1/f*fs$ で求まる.

```python
l2 = int(1/f1*fs)+1
plt.plot(range1[:l2], y1[:l2])
plt.show()
```
