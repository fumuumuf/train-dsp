## 遅延演算 


```python
fs = 8000
t =  np.arange(0,1,1/fs)
s=np.sin(2*np.pi*800*t) + np.sin(2*np.pi*500*t)
cis.audioplay(s,fs)

rg = np.arange(0,100)
plt.subplot(3,1,1)
plt.plot(s[rg])
# 要素の位置をずらすことで遅延処理
# あふれた要素は先頭に挿入される
sd=np.roll(s,5)
plt.subplot(3,1,2)
plt.plot(sd[rg]) 

ssd = s+sd
plt.subplot(3,1,3)
plt.plot(ssd[rg]) 

plt.show()
```
