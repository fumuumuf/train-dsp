```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss
import scipy.fftpack as sfft
import cis
```

```python
# インパルス
imp = np.zeros((1024,))
imp[0] = 1

# 3点移動平均フィルタ
h = np.ones((3,))/3

# インパルス応答
h_imp= np.convolve(h,imp)

plt.stem(h_imp[:100])
```

ナイキスト周波数が1となるようにする

```python
S_imp=sfft.fft(h_imp,1024)
plt.plot(S_imp)
```
