### 色のしきい値で抽出

```python
C = cv2.imread('/root/book_files/image/paprika-966290_960_720.jpg')
C = cv2.cvtColor(C, cv2.COLOR_BGR2RGB)

BW = np.zeros((C.shape[0], C.shape[1]))
BW.shape

# しきい値で制限
BW[np.logical_and(
    np.logical_and(150 < C[:, :, 0], C[:, :, 0] < 240),  # R
    np.logical_and(80 < C[:, :, 1], C[:, :, 1] < 220),   # G
    C[:, :, 2] < 40 # B
)
] = 1

R = C.copy()
R[BW == 0] = 0
plt.imshow(R)
```

テキストでは `np.logical_and`を2引数にしているが, まとめて3つ指定しても大丈夫そうだ.(andなので)
